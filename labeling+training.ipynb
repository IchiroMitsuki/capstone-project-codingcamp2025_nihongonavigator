```python
import csv
import pandas as pd
import requests
from io import StringIO

pd.set_option('display.max_rows', None)
   
# Membaca data kamus kata-kata positif dari GitHub
lexicon_positive = dict()
   
response = requests.get('https://raw.githubusercontent.com/angelmetanosaa/dataset/main/lexicon_positive.csv')
# Mengirim permintaan HTTP untuk mendapatkan file CSV dari GitHub
   
if response.status_code == 200:
    # Jika permintaan berhasil
    reader = csv.reader(StringIO(response.text), delimiter=',')
    # Membaca teks respons sebagai file CSV menggunakan pembaca CSV dengan pemisah koma
   
    for row in reader:
        # Mengulangi setiap baris dalam file CSV
        lexicon_positive[row[0]] = int(row[1])
        # Menambahkan kata-kata positif dan skornya ke dalam kamus lexicon_positive
else:
    print("Failed to fetch positive lexicon data")
   
# Membaca data kamus kata-kata negatif dari GitHub
lexicon_negative = dict()
   
response = requests.get('https://raw.githubusercontent.com/angelmetanosaa/dataset/main/lexicon_negative.csv')
# Mengirim permintaan HTTP untuk mendapatkan file CSV dari GitHub
   
if response.status_code == 200:
    # Jika permintaan berhasil
    reader = csv.reader(StringIO(response.text), delimiter=',')
    # Membaca teks respons sebagai file CSV menggunakan pembaca CSV dengan pemisah koma
   
    for row in reader:
        # Mengulangi setiap baris dalam file CSV
        lexicon_negative[row[0]] = int(row[1])
        # Menambahkan kata-kata negatif dan skornya dalam kamus lexicon_negative
else:
    print("Failed to fetch negative lexicon data")
```

```python
def sentiment_analysis_lexicon_indonesia(
    text, 
    ngram_choices=[1, 2, 3]
):
    """
    Analisis sentimen berbasis lexicon, mendukung frasa dan kata tunggal,
    serta mencegah double counting jika kata sudah termasuk dalam frasa.

    Parameters:
    - text: list of token (hasil tokenisasi)
    - ngram_choices: list panjang n-gram yang ingin diperiksa (default [1,2,3])
    - lexicon_positive: dict kata positif
    - lexicon_negative: dict kata negatif
    - lexicon_positive_phrase: dict frasa positif
    - lexicon_negative_phrase: dict frasa negatif

    Returns:
    - score: skor sentimen
    - polarity: 'positive', 'negative', 'neutral'
    """
    score = 0
    used_indices = set()  # untuk melacak indeks yang sudah termasuk dalam frasa

    # Frasa (n-gram > 1)
    for n in sorted(ngram_choices, reverse=True):  # mulai dari n paling besar
        if n == 1:
            continue  # skip unigram di tahap ini
        for i in range(len(text) - n + 1):
            indices = set(range(i, i + n))
            if used_indices & indices:
                continue  # jika sudah ada bagian frasa yang dipakai, skip
            ngram = ' '.join(text[i:i+n])
            if lexicon_negative_phrase and ngram in lexicon_negative_phrase:
                score -= lexicon_negative_phrase[ngram]
                used_indices.update(indices)
            elif lexicon_positive_phrase and ngram in lexicon_positive_phrase:
                score += lexicon_positive_phrase[ngram]
                used_indices.update(indices)

    # Kata tunggal (n-gram == 1)
    for i, word in enumerate(text):
        if i in used_indices:
            continue  # skip jika sudah dipakai di frasa
        if lexicon_negative and word in lexicon_negative:
            score -= lexicon_negative[word]
        elif lexicon_positive and word in lexicon_positive:
            score += lexicon_positive[word]

    # Tentukan polaritas
    if score > 0:
        polarity = 'positive'
    elif score < 0:
        polarity = 'negative'
    else:
        polarity = 'neutral'

    return score, polarity
```
